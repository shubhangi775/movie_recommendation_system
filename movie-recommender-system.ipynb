```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# We have data about the same movies in two different dataframes. First we combine these two dataframes into one.
```

```python
movies=movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.tail()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+ movies['genres']+ movies['cast']+ movies['crew']
```

```python
new_df=movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# text(tags): vectors
# text vectorization
# Technique: Bag of words
# won't consider stop words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english' )
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
# apply stemming
```

```python
# we will calculate cosine distance and not euclidean distance for vectors
# cosine distance is the angle between two different vectors
# we are working in a 5000-dimensional space. euclidean distance is not a good measure for such a high dimensional space.
# distance is inversely proportional to similarity.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df['title']=='Batman Begins'].index[0]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
